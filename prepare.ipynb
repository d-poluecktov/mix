```python
df = df.drop(
    ['Unnamed: 0', 'Шифр (порядковый номер образца-месяц)', 'Рецептура (цинк/стеарин/бентонит/ЖК, %)', 'М.ч. соли',
     'Массовая доля летучих веществ, %', 'Массовая доля золы, %', 'М100, МПа'], axis=1)
df = df.drop(['А, Шор', 'δ, %'], axis=1)
df
```
